```python
import pandas as pd
```

```python
crash_data = pd.read_csv('/workspaces/TechCatalyst_capston3/tableau/crash dashboard/Motor_Vehicle_Collisions Query.csv')
crash_data.head()
```

```python
crash_data['CRASH TIME'] = pd.to_datetime(crash_data['CRASH TIME'])
```

```python
crash_data['HOUR'] = crash_data['CRASH TIME'].dt.hour
```

```python
crash_data['COLLISION_ID'].count()
```

```python
avg_collisions = crash_data.groupby(['BOROUGH', 'HOUR'])['COLLISION_ID'].count().reset_index()
```

```python
avg_collisions['Percent Increase'] = ((avg_collisions['COLLISION_ID']/47484)*100).round(2)
avg_collisions.columns = ['BOROUGH', 'HOUR', 'Avg Collision Count', 'Percent Increase']
avg_collisions
```

```python
brooklyn = avg_collisions[avg_collisions['BOROUGH'] == 'BROOKLYN']
brooklyn
```

```python
bronx = avg_collisions[avg_collisions['BOROUGH'] == 'BRONX']
bronx
```
